## 1. Objective

The purpose of this notebook is to translate observed usage patterns from the Bellabeat smart device dataset into clear, actionable recommendations for product, marketing, and growth teams.

All conclusions are based on cleaned and validated data prepared in earlier notebooks.


```python
from pathlib import Path
import pandas as pd

project_root = Path.cwd().parents[0]
data_dir = project_root / "data_cleaned"

daily = pd.read_csv(data_dir / "daily_activity_clean.csv", parse_dates=["activity_date"])
sleep = pd.read_csv(data_dir / "sleep_day_clean.csv", parse_dates=["sleep_day"])
hourly_steps = pd.read_csv(data_dir / "hourly_steps_clean.csv", parse_dates=["activity_hour"])

daily.shape, sleep.shape, hourly_steps.shape
```

## 2. Key Behavioural Findings


```python
daily["totalsteps"].describe()
```

### 2.1 Activity Levels

The distribution of daily step counts shows that most users engage in light to moderate activity levels. Median values fall well below high-performance fitness thresholds, while very high activity levels are limited to a small subset of users.

This suggests that the typical Bellabeat user is more focused on maintaining general wellbeing than on intensive fitness training.


```python
sleep.groupby("id")["sleep_day"].count().describe()
```

### 2.2 Sleep Tracking Behaviour

Sleep data is logged inconsistently across users. However, users who do log sleep tend to do so repeatedly, suggesting that sleep tracking is associated with higher overall engagement.

This indicates that sleep tracking may serve as a strong engagement anchor for Bellabeat users.


```python
hourly_steps["hour"] = hourly_steps["activity_hour"].dt.hour
hourly_steps.groupby("hour")["steps"].mean().round(0)
```

### 2.3 Daily Usage Routines

Average step counts peak during morning, midday, and early evening hours. This pattern suggests that engagement with the device is driven by predictable daily routines rather than spontaneous activity.

These habitual usage windows are critical for timing notifications and in-app prompts.


## 3. Strategic Conclusions

Based on the analysis, several high-level conclusions emerge:

- The Bellabeat user base is primarily wellness-oriented rather than performance-focused
- Engagement aligns strongly with daily routines
- Sleep tracking correlates with higher overall engagement
- A small highly active user segment exists but is not representative of the majority


## 4. Recommendations for Product & Marketing Teams

The following recommendations translate these behavioural patterns into practical actions that Bellabeatâ€™s product and marketing teams could implement to improve engagement and retention.

### 1. Position Bellabeat as a daily wellness companion  
Emphasise sustainable habits, balance, and everyday movement rather than athletic performance.

### 2. Implement adaptive goal-setting  
Use personalised goals based on recent activity levels to encourage consistency without intimidation.

### 3. Strengthen sleep-focused engagement  
Encourage regular sleep logging and provide meaningful insights linked to recovery and wellbeing.

### 4. Segment users and personalise experiences  
Tailor content and notifications for low, moderate, and high activity users.

### 5. Time nudges around natural activity peaks  
Align notifications with morning, midday, and early evening routines.

### 6. Identify disengagement risk early  
Use low activity and missing sleep data as early signals for supportive re-engagement.

### 7. Reward habit formation  
Celebrate streaks and consistency rather than extreme performance.
